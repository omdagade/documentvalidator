```python
from PIL import Image
import pytesseract
import numpy as np
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'
```

```python
import os
from os import listdir
import cv2
```

```python
folder_dir = "pandata"
pandata=[]
for images in os.listdir(folder_dir):
    
    
    img = cv2.imread(os.path.join(folder_dir,images))
    if img is not None:
         
        img1=np.array(img)
        text=pytesseract.image_to_string(img1)
        pandata.append(text)
        

    
```

```python
pandata
```

```python
import pandas as pd
```

```python
import string
```

```python
pandatanew=[]
for i in pandata:
    new_string = i.translate(str.maketrans('', '', string.punctuation))
    
    pandatanew.append(new_string)
```

```python
pandatanew
```

```python
import nltk
nltk.download('punkt')
```

```python
tokenized=[]
for i in pandatanew:
    nltk_tokens = nltk.word_tokenize(i)
    tokenized.append(nltk_tokens)
```

```python
tokenized
```

```python
pandf=pd.DataFrame()
for i in tokenized:
    tempdict={}
    for j in i:
        a=j.upper()
        if a in tempdict:
            tempdict[a]=tempdict[a]+1
        else:
            
            tempdict[a]=1
   
    pandf = pd.concat([pandf, pd.DataFrame([tempdict])], ignore_index=True)
```

```python
pandf
```

```python
pandf=pandf.fillna(0)
```

```python
pandf
```

```python
pandf["VALID"]=1
```

```python
pandf
```

```python
folder_dir = "otherdoc"
otherdata=[]
for images in os.listdir(folder_dir):
    
    
    img = cv2.imread(os.path.join(folder_dir,images))
    if img is not None:
         
        img1=np.array(img)
        text=pytesseract.image_to_string(img1)
        otherdata.append(text)

otherdatanew=[]
for i in otherdata:
    new_string = i.translate(str.maketrans('', '', string.punctuation))
    
    otherdatanew.append(new_string)

tokenized=[]
for i in otherdatanew:
    nltk_tokens = nltk.word_tokenize(i)
    tokenized.append(nltk_tokens)

otherdf=pd.DataFrame()
for i in tokenized:
    tempdict={}
    for j in i:
        a=j.upper()
        if a in tempdict:
            tempdict[a]=tempdict[a]+1
        else:
            
            tempdict[a]=1
   
    otherdf = pd.concat([otherdf, pd.DataFrame([tempdict])], ignore_index=True)

        

    
```

```python
otherdf
```

```python
otherdf=otherdf.fillna(0)
```

```python

otherdf["VALID"]=0
```

```python
otherdf
```

```python
dataset= pd.concat([pandf, otherdf], ignore_index=True)
```

```python
dataset["VALID"]
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.fillna(0)
```

```python
dataset.isnull().sum()
```

```python
pd.set_option('display.max_columns', None)
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
Y=dataset["VALID"]
X=dataset[["INCOME","TAX","DEPARTMENT","GOVT","OF","INDIA","PERMANENT", "NUMBER", "NAME" , "FATHERS", "BIRTH"]]
```

```python
Y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30) 
```

```python
clf = RandomForestClassifier(n_estimators = 100)   
clf.fit(X_train, y_train) 
y_pred = clf.predict(X_test) 
```

```python
result = pd.DataFrame()
result["actual"] = y_test
result["predictions"] = y_pred
```

```python
result
```

```python
from sklearn import metrics 
```

```python
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred)) 
```
